```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import json
from pyproj import Proj, transform
import warnings
from scipy.interpolate import griddata
```

```python
em_df = pd.read_csv("DWR_Survey5/em_resistivity.csv")
thickness_df= pd.read_csv("DWR_Survey5/thickness.csv")
def read_AEM_Data_redivis(df, thickness_df):
    #transform coordinates
    output_crs = 'epsg:32611' # Coordinate system of the file
    input_crs = 'epsg:3310' # Global lat-lon coordinate system
    x, y = transform(input_crs, output_crs, df.UTMX.values, df.UTMY.values)
    df['UTMX'], df['UTMY'] = x, y
    #sort df
    lines = df.LINE_NO.unique()
    df_list = []
    for l in lines:
        df_temp = df[df.LINE_NO==l]
        df_list.append(df_temp.sort_values(['UTMX', 'UTMY'], ascending=[1, 0]))
    df = pd.concat(df_list)
    topo = df[['UTMX', 'UTMY', 'ELEVATION']].values[:, :]
    line = df['LINE_NO'].values
    hz = np.array(thickness_df.values[0][0][1:-1].split(',')).astype(float)
    hz = np.r_[hz, hz[-1]]
    res_dics = df.MEASUREMENTS.values
    n_sounding, n_layers = len(res_dics), len(hz)
    resistivity = np.zeros([n_sounding, n_layers])
    for i in range(n_sounding):
        rho = json.loads(res_dics[i])['RHO']
        resistivity[i] = rho
    return df, topo, hz, resistivity, line
    
df_SA5, topo_SA5, hz_SA5, resistivity_SA5, line_SA5 = read_AEM_Data_redivis(em_df, thickness_df)
```

```python
AEM_extent = [254290.296875, 257898.921875, 4089789.0, 4093032.25]
def read_AEM_Data_csv(filename):
    df = pd.read_csv(filename)
    header = df.columns
    tmp_rho = np.array(header)[['RHO' in header for header in header]]
    rho_names = tmp_rho[:int(len(hz_SA5))]
    output_crs = 'epsg:32611' # Coordinate system of the file
    input_crs = 'epsg:4326' # Global lat-lon coordinate system
    x, y = transform(input_crs, output_crs, df.LATITUDE.values, df.LONGITUDE.values)
    df['UTMX'], df['UTMY'] = x, y
    df = df.query("254291<UTMX<257899")
    df = df.query("4089789<UTMY<4093032")
    topo = df[['UTMX', 'UTMY', 'ELEVATION']].values[:, :]
    line = np.arange(len(df.UTMX.values))
    temp_line = 0*line
    temp_line[line<57]=1
    df['LINE_NO']=temp_line
    hz = hz_SA5
    resistivity= df[rho_names].values[:,:]
    line = df['LINE_NO'].values
    return df, topo, hz, resistivity, line

df_foothills, topo_foothills, hz_foothills, resistivity_foothills, line_foothills = read_AEM_Data_csv("resistivity_tulare_kaweah_foothill_latest.csv")
```

```python
plt.scatter(topo_SA5[:,0], topo_SA5[:,1], c = resistivity_SA5[:,2], vmin = 0, vmax = 150, s = 5)
plt.scatter(topo_foothills[:,0], topo_foothills[:,1], c = resistivity_foothills[:,2], vmin = 0, vmax = 150, s = 5)
```

```python
n_SA5 = resistivity_SA5.shape[0]
n_foothill = resistivity_foothills.shape[0]
n = n_SA5+n_foothill

resistivity = np.zeros([n, 30])
resistivity[0:n_SA5] = resistivity_SA5
resistivity[n_SA5:] = resistivity_foothills

topo = np.zeros([n, 3])
topo[0:n_SA5]=topo_SA5
topo[n_SA5:]=topo_foothills

line = np.append(line_SA5, line_foothills)

DOI = np.append(df_SA5.DOI_STANDARD.values, df_foothills.DOI_STANDARD.values)

sounding = np.arange(n)
dic = {"sounding":sounding, "X":topo[:,0], "Y":topo[:,1], "Z":topo[:,2], "line":line, "DOI":DOI}

df = pd.DataFrame(dic)
```
